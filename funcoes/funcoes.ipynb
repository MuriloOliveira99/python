# Funções em Python

#### Função para personalizar um título

```python
def titulo(txt):
  print('-' * int(len(txt)+1))
  print(txt)
  print('-' * int(len(txt)+1))

# Chamando a função e passando o parametro.
titulo('APRENDENDO FUNÇÕES EM PYTHON')
titulo('MURILO OLIVEIRA')
```

# Função para contar a quantidade de números passados por parametros.<br> E também aplicando o empacotamento

```python
# * = significa que a função vai receber uma qtd. indefinida de parametros
# Retorna uma tupla dos valores
def contador(*num):
  tamanho = len(num)
  print(f'Recebi os valores {num} e ao todo são {tamanho} números.')

contador(2, 1, 7)
```

# Função para dobrar os valores de uma lista

```python
def dobrar_valor(lista):
  posicao = 0
  while posicao < len(lista):
    lista[posicao]*=2
    posicao+=1

valores = [10, 20, 30]
dobrar_valor(valores)
print(valores)
```

#### 1. Faça um programa que tenha uma função chamada area, que receba as dimensões de um terreno retangular(L x C) e mostre a área do terreno.

```python
def area(largura = 0, comprimento = 0):
  return largura * comprimento

largura = float(input('Largura: '))
comprimento = float(input('Comprimento: '))
calculo = area(largura, comprimento)

print(f'A área de um terreno {largura} x {comprimento} é {calculo}m²')
```

#### 3. Faça um programa que tenha uma função chamada contador, que receba três parâmetros: inicio, fim e passo e realiza a contagem.<br>Seu programa tem que realizar três contagens atráves da função criada:<br>a) de 1 até 10, de 1 em 1 <br>b) De 10 até 0, de 2 em 2 <br>c) uma contagem personalizada.

```python
from time import sleep

def contador(inicio, fim, passo):
  if passo == 0:
    passo = 1
    
  if inicio < fim:
    for i in range(inicio, fim+1, passo):
      print(i, end=' ')
      sleep(1)
  elif inicio > fim:
    for i in range(inicio, fim-1, -passo):
      print(i, end=' ')
      sleep(1)
  else:
    print('Iguais')

print('Contagem de 1 até 10, de 1 em 1.')
for a in range(1, 11, 1):
  sleep(1)
  print(a, end=' ')

print('\n')

print('Contagem de 10 até 0, de 2 em 2.')
for b in range(10, -2, -2): 
  sleep(1)
  print(b, end=' ')

print('\n\nAgora é sua vez de personalizar a contagem!\n')

inicio = int(input('Início: '))
fim = int(input('Fim: '))
passo = abs(int(input('Passo: ')))

print(f'\nContagem de {inicio} até {fim}, de {passo} em {passo}')
contador(inicio, fim, passo)
```

#### 5. Faça um programa que tenha uma lista chamada números e duas funções chamadas sorteia e somaPar. A primeira função vai sortear 5 números e vai colocá-los dentro da lista e a segunda função vai mostrar a soma entre todos os valores PARES sorteador pela função anterior.

```python
from random import randint
numeros = []

def sorteia():
  for i in range(5):
    rand_num = randint(0, 1000)
    numeros.append(rand_num)
  
def soma_par(lista):
  soma = 0
  print('Números pares:')
  for num in lista:
    if num % 2 == 0:
      print(num, end=' ')
      soma = soma + num
  return soma 

sorteia()
soma = soma_par(numeros)
print(f'\n\nNúmeros gerados aleatoriamente: \n{numeros}\n')
print(f'A soma dos números pares é: {soma}')
```

#### 6. Crie um programa que tenha uma função chamada voto(), que vai receber como parâmetro o ano de nascimento de uma pessoa. retornando um valor literal indicando se uma pessoa tem voto NEGADO, OPCIONAL ou OBRIGATÓRIO nas eleições

```python
import datetime

def voto(ano_nascimento):
  ano_atual = datetime.date.today().year
  idade = ano_atual - ano_nascimento 

  if 18 <= idade <= 70:
      obrigatoriedade = 'VOTO OBRIGATÓRIO'
  elif idade < 16:
      obrigatoriedade = 'NÃO VOTA'
  else:
      obrigatoriedade = 'VOTO OPCIONAL'
  print(f'Com {idade} anos: {obrigatoriedade}.')
  

ano_nascimento = int(input('Ano de nascimento: '))

voto(ano_nascimento)
```
