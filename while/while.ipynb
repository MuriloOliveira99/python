# WHILE

# Exemplo com FOR 

```python
# for -> quando sabemos o "limite"
for c in range(1, 10):
  print(c)
```

# Exemplo WHILE

```python
# while -> quando não sabemos o "limite".
c = 1
while c < 10:
  print(c)
  c+=1 #c = c +1
```

```python
# Variavel contadora. Para controlar a quantidade de vezes que eu quero exibir a mensagem 'Teste'. Ela inicia em 1
qtd_mensagem = 1

# Enquanto a variavel qtd_mensagem for MENOR ou IGUAL a 10, exibe a mensagem 'teste' 10 VEZES
while qtd_mensagem <=10:
  print('Teste')
  qtd_mensagem = qtd_mensagem+1 # pega o valor inicial da variavel qtd_mensagem e soma MAIS UM
```

```python
contadora = 0
acumuladora = 0

while contadora <= 10: # Quantas números eu quero que o usuario digite?
  numero = int(input('Digite um número; '))
  contadora = contadora + 1 # Pega o valor inicial da variavel contadora e soma mais 1 até chegar no 10
  acumuladora = acumuladora + numero # Pega o valor inicial da variável contadora e soma mais o número informado pelo usuario
print(contadora, acumuladora) 
```

# Exemplo WHILE - Quantidade números pares/ímpares

```python
n = 1
par = impar = 0
while n != 0:
  n = int(input("Digite um valor: "))
  if n!=0: # remove a flag do resultado final
    if n % 2 == 0:
      par += 1
    else:
      impar += 1

print(f'Você digitou {par} número(s) par(es) e  {impar} número(s) ímpar(es)')
```

#### 9. Faça um programa que peça 10 números inteiros, calcule e mostre a quantidade de números pares e a quantidade de números impares.

```python
par = 0
impar = 0
lista_num = []

while len(lista_num) < 10:
  num = int(input('Digite um número: '))
  lista_num.append(num)

  if num % 2 == 0:
    par+=1
  else:
    impar+=1

print(f'\nNúmeros digitados: {lista_num}')
print(f'Qtd. números par(es): {par}')
print(f'Qtd. números impar(es): {impar}')
```

#### 11. Faça um programa que, dado um conjunto de N números, determine o menor valor, o maior valor, a soma dos valores e aceite apenas números entre 0 e 1000.

```python
lista_numeros = []

while True:
  numero = int(input('Número(s): '))
  if numero < 0 or numero > 1000:
    break
  lista_numeros.append(numero)

print(f'\nO menor número é: {min(lista_numeros)}')
print(f'O maior número é: {max(lista_numeros)}')
print(f'A soma dos números é: {sum(lista_numeros)}')
```

#### 12. O computador vai "pensar" em um número entre 0 e 10. Só que agora o jogador vai tentar adivinhar até acertar, mostrando no final quantos palpites foram necessários para vencer

```python
import random

computador = random.randint(0, 10)
jogador = int(input('Digite um número: '))
cont_palpites = 0

while computador != jogador:
  jogador = int(input('Digite um número: '))
  cont_palpites+=1

print(f'Parabéns! O número do Computador era {computador}')
print(f'Qtd. de Chances: {cont_palpites}')
```

#### 14. Faça um programa que leia um número qualquer e mostre o seu fatorial. <br> Exemplo: 5x4x3x2x1=120

```python
numero = int(input('Digite um número: '))
fatorial = numero
cont = 1
while fatorial > 0:
  print(f'{fatorial}')
  cont = cont * fatorial
  fatorial = fatorial - 1
print(cont)
```
